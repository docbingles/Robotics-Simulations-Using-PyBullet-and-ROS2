# **Importing Libraries**

```python
import pybullet as p
import pybullet_data
import numpy as np
import matplotlib.pyplot as plt
```

# **Defining Helper Functions**

```python
def write_urdf(filename, content):
    with open(filename, 'w') as f:
        f.write(content)
    print(f"{filename} saved.")

def render(cam_pos, target_pos, width=512, height=512):
    view_matrix = p.computeViewMatrix(cameraEyePosition=cam_pos,
                                     cameraTargetPosition=target_pos,
                                     cameraUpVector=[0, 0, 1])
    proj_matrix = p.computeProjectionMatrixFOV(fov=60,
                                               aspect=width/height,
                                               nearVal=0.1,
                                               farVal=5)
    _, _, rgba, _, _ = p.getCameraImage(width, height,
                                        view_matrix, proj_matrix,
                                        renderer=p.ER_BULLET_HARDWARE_OPENGL)
    img = np.reshape(rgba, (height, width, 4))[:, :, :3]
    return img
```

# **Creating and Saving the Box Table Model**

```python
box_table_urdf = """
<robot name="box_table">
  <link name="link">
    <inertial>
      <origin xyz="0 0 0"/>
      <mass value="2"/>
      <inertia ixx="0.2" ixy="0" ixz="0" iyy="0.2" iyz="0" izz="0.2"/>
    </inertial>
    <visual>
      <origin xyz="0 0 0"/>
      <geometry>
        <box size="1.2 0.8 0.5"/>
      </geometry>
      <material name="wood_brown">
        <color rgba="0.55 0.27 0.07 1"/>
      </material>
    </visual>
    <collision>
      <origin xyz="0 0 0"/>
      <geometry>
        <box size="1.2 0.8 0.5"/>
      </geometry>
    </collision>
  </link>
</robot>
"""

write_urdf('box_table.urdf', box_table_urdf)
```

# **Visualizing the Box Table in PyBullet**

```python
cid = p.connect(p.DIRECT)
p.setAdditionalSearchPath(pybullet_data.getDataPath())

p.loadURDF("plane.urdf")

body = p.loadURDF('box_table.urdf', basePosition=[0, 0, 0.25])

img = render(cam_pos=[1, 1, 1], target_pos=[0, 0, 0.25])

plt.imshow(img)
plt.axis('off')
plt.title("Visualizing Box Table")
plt.show()

p.disconnect()
```

# **Rendering the Scene from Multiple Camera Angles**

```python
camera_positions = [
    [2, 2, 2.5],
    [0, -3, 2.2],
    [-3, 1, 2.5],
]

target_position = [0, 0, 0.5]

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

for i, cam_pos in enumerate(camera_positions):
    img = render(cam_pos, target_position)
    axes[i].imshow(img)
    axes[i].axis('off')
    axes[i].set_title(f"View {i+1}: Cam Pos {cam_pos}")

plt.suptitle("PyBullet Scene from Multiple Camera Views", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
